```python
import requests
import pandas as pd

# API من Open Food Facts (نجيب 100 منتج غذائي)
url = "https://world.openfoodfacts.org/cgi/search.pl"
params = {
    "action": "process",
    "json": 1,
    "page_size": 100,   # عدد المنتجات
    "page": 1
}

response = requests.get(url, params=params)
data = response.json()

# استخراج بيانات أساسية
products = []
for item in data.get("products", []):
    name = item.get("product_name", "N/A")
    brand = item.get("brands", "N/A")
    category = item.get("categories", "N/A")
    quantity = item.get("quantity", "N/A")
    products.append([brand, name, category, quantity])

# تحويل لجدول
df = pd.DataFrame(products, columns=["الماركة", "اسم المنتج", "الفئة", "الكمية"])

# حفظ في CSV
df.to_csv("food_products.csv", index=False, encoding="utf-8-sig")

# عرض أول 20 منتج
df.head(20)
```
