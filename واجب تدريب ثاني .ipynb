```python
import requests
import pandas as pd

url = "https://www.googleapis.com/books/v1/volumes"

all_books = []

# نعمل Loop على صفحات النتائج (كل مرة max 40 كتاب)
for start in range(0, 200, 40):  # هنا بنجيب 200 كتاب (تقدر تزود)
    params = {
        "q": "comic books",    # الكلمة المفتاحية
        "maxResults": 40,      # الحد الأقصى المسموح
        "startIndex": start,
        "printType": "books"
    }
    
    response = requests.get(url, params=params)
    data = response.json()
    
    for item in data.get("items", []):
        title = item["volumeInfo"].get("title", "N/A")
        sale_info = item.get("saleInfo", {})
        price = "غير متوفر"
        if "retailPrice" in sale_info:
            price = f"{sale_info['retailPrice']['amount']} {sale_info['retailPrice']['currencyCode']}"
        all_books.append([title, price])

# تحويل لجدول DataFrame
df = pd.DataFrame(all_books, columns=["اسم القصة", "السعر"])

# حفظ الجدول في CSV كقاعدة بيانات
df.to_csv("comic_books.csv", index=False, encoding="utf-8-sig")

df.head(20)  # عرض أول 20 كتاب كمثال
```
